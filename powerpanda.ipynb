```python
import pandapower as pp
import pandapower.contingency as pc
```

```python
net = pp.from_json("PowerMCP/pandapower/test_case.json")

# Define the base thermal limits so the contingency algorithm has a threshold to evaluate
net.line['max_loading_percent'] = 100.0

# If your network includes transformers, you must set their limits as well
if 'trafo' in net and not net.trafo.empty:
    net.trafo['max_loading_percent'] = 100.0

# 2. Define your N-1 cases 
# You pass a dictionary specifying which elements to take offline.
nminus1_cases = {
    "line": {"index": net.line.index.values},
    "trafo": {"index": net.trafo.index.values}
}

# 3. Run the contingency analysis
# This evaluates the base case (N-0) and all defined N-1 cases
pc.run_contingency(net, nminus1_cases)

# 4. View the worst-case aggregated results
# The function appends new columns to your standard result tables
print("Line Results:")
print(net.res_line[['loading_percent', 'max_loading_percent', 'cause_element', 'cause_index']])

print("\nBus Results:")
print(net.res_bus[['vm_pu', 'min_vm_pu', 'max_vm_pu']])
```
